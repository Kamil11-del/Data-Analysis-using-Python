## 📌 Section 1: Variables and Data Types

### What is a Variable?

Think of a **variable** as a **labelled box** where you store a value.

```
  ┌─────────────┐
  │     25      │  ← value stored inside
  └─────────────┘
       age         ← label (variable name)
```

- You give the box a **name** (the variable name)
- You put a **value** inside it
- Later, you can **use the name** to get the value back
- You can also **change** the value at any time

### Rules for Naming Variables:
| ✅ Valid | ❌ Invalid | Why Invalid |
|---------|-----------|-------------|
| `student_name` | `student name` | No spaces allowed |
| `age25` | `25age` | Cannot start with a number |
| `_score` | `@score` | Only letters, numbers, underscore |
| `totalMarks` | `total-marks` | No hyphens |

> 🏆 **Best Practice:** Use `snake_case` — all lowercase with underscores. Example: `student_grade`, `total_salary`

````python
# ── Creating Your First Variables ──────────────────────────────────────────
# The '=' sign means ASSIGNMENT — "store this value in this variable"
# It does NOT mean 'equal to' like in mathematics!
# these below are variables

student_name = "Ahmed"
student_age  = 22
city         = "Islamabad" 

# Now let's USE those variables
print(student_name)
print(student_age)
print(city)
````

### 🔷 Data Types — The 4 Essential Types

Every value in Python has a **type**. The type tells Python what kind of data it is and what operations are allowed.

| Type | Full Name | Examples | Real-World Use |
|------|-----------|---------|----------------|
| `int` | Integer | `25`, `-3`, `1000` | Age, count, year |
| `float` | Floating Point | `3.14`, `99.9`, `-0.5` | Price, temperature, percentage |
| `str` | String | `"Hello"`, `'Ahmed'` | Name, city, any text |
| `bool` | Boolean | `True`, `False` | Yes/No, On/Off, Pass/Fail |

> 🔑 **Why does type matter?**  
> You can add two numbers: `5 + 3 = 8`  
> You can join two strings: `"Hello" + " World" = "Hello World"`  
> But you **cannot** mix them without converting: `5 + "3"` → ❌ ERROR

## 📌 Section 2: The `print()` Function and f-strings

### The `print()` Function

`print()` is how Python **shows output** to the user. It's the most used function in Python.

You've already been using it! Now let's understand it deeply.

### f-strings — The Modern Way to Format Output

An **f-string** (formatted string) lets you **embed variables directly inside a string**.

- Put `f` before the opening quote
- Put variable names inside `{ }` curly braces

```python
name = "Ahmed"
age = 22
print(f"Hello, {name}! You are {age} years old.")  ← f-string
```

**Output:** `Hello, Ahmed! You are 22 years old.`

> 🏆 f-strings are the **recommended** way in Python 3.6+. Always prefer them.

````python
name="Ahmed"
age = 22
print("Hello"+name+" You are age is :",age)
````

````python
print(f"Hello {name} You are age is :{age}")
````

````python
# ── print() Basics ─────────────────────────────────────────────────────────

# Simple text
print("Hello, World!")

# Multiple items (separated by comma) — print adds a space between them
print("Name:", "Ahmed", "Age:", 22)

# Empty print = blank line
print()

# Custom separator (sep) — replaces the default space
print("\n2024", "03", "15", sep="-")    # Output: 2024-03-15
print("Ahmed", "Khan", sep="_")        # Output: Ahmed_Khan

# End parameter — by default print adds a newline at end
print("Loading", end="...")
print("Done!")   # appears on SAME line as Loading
````

````python
# ── Old way vs f-string way ────────────────────────────────────────────────

name     = "Fatima"
age      = 20
gpa      = 3.75
city     = "Lahore"

# ❌ Old way — concatenation (messy, error-prone)
print("Name: " + name + ", Age: " + str(age) + ", GPA: " + str(gpa))

# ✅ f-string — clean and readable
print(f"Name: {name}, Age: {age}, GPA: {gpa}")

# ✅ f-string with a whole sentence
print(f"Hello! I'm {name}, {age} years old from {city}.")
print(f"My GPA is {gpa} and I'm studying Data Analytics.")
````

````python
# ── f-string: Expressions Inside {} ───────────────────────────────────────
# You can put ANY Python expression inside the curly braces — not just variables!

price    = 1200
quantity = 3

# Calculations INSIDE the f-string
print(f"Unit price: PKR {price}")
print(f"Quantity: {quantity}")
print(f"Total: PKR {price * quantity}")   # calculation inside {}

# Calling functions inside {}
text = "hello world"
print(f"Original: {text}")
print(f"Uppercase: {text.upper()}")
print(f"Length: {len(text)} characters")
````

````python
num1 = 2
num2 =4
sum = num1+num2
print(f"The total of both numbers is: {num1+num2}")
````

````python
# ── f-string: Number Formatting ────────────────────────────────────────────
# Control how numbers are displayed using format specifiers
# Syntax: {variable:.Xf}  where X is number of decimal places

pi          = 3.14159265358979
salary      = 85000
percentage  = 87.333333

# Decimal places
print(f"Pi (2 decimal places): {pi:.2f}")
print(f"Pi (4 decimal places): {pi:.4f}")

# Percentage sign
rate = 0.875
print(f"Success rate: {rate:.1%}")        # converts to percentage automatically

# Thousands separator (comma)
print(f"Monthly salary: PKR {salary:,}")

# Combined
print(f"Score: {percentage:.1f}%")
````

## 📌 Section 3: Basic Arithmetic Operations

Python is an excellent calculator. Here are all the arithmetic operators:

| Operator | Symbol | Example | Result | Description |
|----------|--------|---------|--------|-------------|
| Addition | `+` | `10 + 3` | `13` | Adds two numbers |
| Subtraction | `-` | `10 - 3` | `7` | Subtracts second from first |
| Multiplication | `*` | `10 * 3` | `30` | Multiplies two numbers |
| Division | `/` | `10 / 3` | `3.333...` | Always returns a float |
| Floor Division | `//` | `10 // 3` | `3` | Divides, rounds DOWN to int |
| Modulus | `%` | `10 % 3` | `1` | Remainder after division |
| Exponent | `**` | `10 ** 3` | `1000` | 10 to the power of 3 |

> 🔑 **Order of operations (PEMDAS):** Python follows standard math rules:
> `**` → `*` `/` `//` `%` → `+` `-`  
> Use **parentheses** `()` to control the order!

````python
# ── All Arithmetic Operators ────────────────────────────────────────────────

a = 17
b = 5

print(f"a = {a},  b = {b}")
print("─" * 30)
print(f"Addition       : {a} + {b} = {a + b}")
print(f"Subtraction    : {a} - {b} = {a - b}")
print(f"Multiplication : {a} * {b} = {a * b}")
print(f"Division       : {a} / {b} = {a / b}")
print(f"Floor Division : {a} // {b} = {a // b}")   # result is 3, not 3.4
print(f"Modulus        : {a} % {b} = {a % b}")     # 17 = 5*3 + 2, so remainder is 2
print(f"Exponent       : {a} ** 2 = {a ** 2}")    # 17 squared
````

````python
# ── Floor Division vs Regular Division ────────────────────────────────────
# This is an important concept!

total_students = 47
per_group      = 5

complete_groups  = total_students // per_group   # how many FULL groups
leftover         = total_students % per_group    # how many students left over

print(f"Total students    : {total_students}")
print(f"Students per group: {per_group}")
print(f"Complete groups   : {complete_groups}")
print(f"Students left over: {leftover}")
print()

# Modulus real-world use: check if even or odd
number = 28
if number % 2 == 0:
    print(f"{number} is EVEN")
else:
    print(f"{number} is ODD")
````

````python
result = 7//3
print(result)
````

````python
# ── Order of Operations ─────────────────────────────────────────────────────
# Python follows PEMDAS — same as mathematics

# Without parentheses
result1 = 10 + 5 * 2    # multiplication first: 10 + 10 = 20
print(f"10 + 5 * 2       = {result1}")   # = 20, NOT 30

# With parentheses
result2 = (10 + 5) * 2   # parentheses first: 15 * 2 = 30
print(f"(10 + 5) * 2     = {result2}")   # = 30

# Complex expression
result3 = 2 ** 3 + 10 / 2 - 1
# Step 1: 2**3 = 8
# Step 2: 10/2 = 5.0
# Step 3: 8 + 5.0 - 1 = 12.0
print(f"2 ** 3 + 10 / 2 - 1 = {result3}")

# Always use parentheses when in doubt — it makes code READABLE
result4 = (2 ** 3) + (10 / 2) - 1  # same result, more readable
print(f"(2**3) + (10/2) - 1  = {result4}")
````

````python
# ── Shorthand Assignment Operators ─────────────────────────────────────────
# These are shortcuts — very common in real code!

score = 100
print(f"Starting score: {score}")

score += 10     # same as: score = score + 10
print(f"After += 10  : {score}")

score -= 5      # same as: score = score - 5
print(f"After -= 5   : {score}")

score *= 2      # same as: score = score * 2
print(f"After *= 2   : {score}")

score //= 3     # same as: score = score // 3
print(f"After //= 3  : {score}")
````

````python
# ── Useful Built-in Math Functions ─────────────────────────────────────────

# abs() — absolute value (removes negative sign)
print(f"abs(-45)   = {abs(-45)}")
print(f"abs(7.3)   = {abs(7.3)}")

# round() — round to N decimal places
pi = 3.14159
print(f"round({pi}, 2) = {round(pi, 2)}")
print(f"round({pi}, 0) = {round(pi, 0)}")

# max() and min()
marks = 88
print(f"\nmax(45, 88, 72, 91) = {max(45, 88, 72, 91)}")
print(f"min(45, 88, 72, 91) = {min(45, 88, 72, 91)}")

# pow() — power (same as **)
print(f"\npow(2, 10) = {pow(2, 10)}")

# For more math: import the math module
import math
print(f"\nmath.sqrt(144) = {math.sqrt(144)}")
print(f"math.pi        = {math.pi:.6f}")
print(f"math.ceil(4.1) = {math.ceil(4.1)}")   # round UP always
print(f"math.floor(4.9)= {math.floor(4.9)}")  # round DOWN always
````

## 📌 Section 4: Comments and Documentation

### What are Comments?

Comments are notes written **for humans** — Python completely ignores them when running your code.

### Why Comments Matter?

Imagine you wrote code 6 months ago. Without comments, you won't remember what anything does!  
Comments help:
- **You** understand your own code later
- **Your team** understand your code
- **Others** who use or review your code

> 💬 *"Code tells you HOW. Comments tell you WHY."*

### Types of Comments in Python:

```python
# Single-line comment — starts with #

"""  
Multi-line comment (docstring)
Used to describe functions, classes, or modules
"""
```

````python
# ── Single-Line Comments ────────────────────────────────────────────────────

# This entire line is a comment — Python skips it
"""djfljlkdasjfjlkadsjlf
flkdsjfjldskjflk;;dsjfljads
ldnflkdsjklfjadskljflkdsjaf"""
student_age = 22   # This is an inline comment — code runs, then comment is ignored

# Comments can explain what your code does
total_marks = 450
max_marks   = 500

# Calculate percentage — divide actual marks by total possible
percentage = (total_marks / max_marks) * 100

# Display result with 2 decimal places
print(f"Percentage: {percentage:.2f}%")
````

````python
# ── Multi-line Comments (Docstrings) ───────────────────────────────────────
# Use triple quotes ''' or """
# Main use: documenting functions, scripts, notebooks

"""
This script calculates a student's final grade.

Author  : Ahmed Khan
Date    : 2024-03-15
Version : 1.0

Inputs  : Quiz marks, Assignment marks, Final exam marks
Output  : Final percentage and letter grade
"""

# The docstring above is skipped by Python but is good documentation practice
print("Script documentation is in the docstring above this line.")
````

````python
# ── Good vs Bad Comments ───────────────────────────────────────────────────

price    = 500
discount = 0.15

# ❌ BAD comment: just restates what the code says — useless!
# multiply price by discount
savings = price * discount

# ✅ GOOD comment: explains WHY or the business logic
# Apply 15% seasonal discount for Eid sale
savings = price * discount
final_price = price - savings

print(f"Original price : PKR {price}")
print(f"Discount (15%) : PKR {savings}")
print(f"Final price    : PKR {final_price}")

# ✅ Use comments to temporarily disable code (useful for debugging!)
# print("This line is disabled")
print("This line runs normally")
````

````python
# ══════════════════════════════════════════════════════════════════
#  🧮 MINI-PROJECT: SIMPLE CALCULATOR
# ══════════════════════════════════════════════════════════════════

"""
Simple Calculator
Performs: addition, subtraction, multiplication, division,
          floor division, modulus, and exponent
"""

# ─── STEP 1: Get input from user ───────────────────────────────────
print("=" * 40)
print("      🧮 SIMPLE PYTHON CALCULATOR")
print("=" * 40)

# input() returns a string — float() converts it to a number
num1 = float(input("Enter first number  : "))
num2 = float(input("Enter second number : "))

# ─── STEP 2: Perform calculations ──────────────────────────────────
addition       = num1 + num2
subtraction    = num1 - num2
multiplication = num1 * num2
division       = num1 / num2   if num2 != 0 else "Error: Division by zero!"
floor_div      = num1 // num2  if num2 != 0 else "Error: Division by zero!"
modulus        = num1 % num2   if num2 != 0 else "Error: Division by zero!"
power          = num1 ** num2

# ─── STEP 3: Display results ────────────────────────────────────────
print()
print(f"  Numbers entered: {num1} and {num2}")
print("─" * 40)
print(f"  Addition        : {num1} + {num2} = {addition}")
print(f"  Subtraction     : {num1} - {num2} = {subtraction}")
print(f"  Multiplication  : {num1} x {num2} = {multiplication}")
print(f"  Division        : {num1} ÷ {num2} = {division}")
print(f"  Floor Division  : {num1} // {num2} = {floor_div}")
print(f"  Modulus         : {num1} % {num2} = {modulus}")
print(f"  Power           : {num1} ^ {num2} = {power}")
print("=" * 40)
print("  ✅ Calculation complete!")
print("=" * 40)
````

````python
# ══════════════════════════════════════════════════════════════════
#  💰 PRACTICE PROJECT: MONTHLY BUDGET CALCULATOR
# ══════════════════════════════════════════════════════════════════

"""
Monthly Budget Calculator
Helps you plan your monthly expenses based on your salary.
Budget breakdown:
  - Savings      : 20%
  - Rent         : 30%
  - Food         : 20%
  - Transport    : 10%
  - Utilities    : 10%
  - Entertainment: 5%
  - Emergency    : 5%
"""

# ─── STEP 1: Get salary from user ──────────────────────────────────
print("=" * 50)
print("MONTHLY BUDGET CALCULATOR")
print("=" * 50)

monthly_salary = float(input("Enter your monthly salary (PKR): "))

# ─── STEP 2: Define budget percentages ─────────────────────────────
# Using variables for percentages makes it easy to change later
savings_pct       = 0.20   # 20% for savings
rent_pct          = 0.30   # 30% for rent/housing
food_pct          = 0.20   # 20% for food & groceries
transport_pct     = 0.10   # 10% for transportation
utilities_pct     = 0.10   # 10% for electricity, internet, water
entertainment_pct = 0.05   # 5% for leisure
emergency_pct     = 0.05   # 5% emergency fund

# Quick check: percentages must add up to 100%
total_pct = (savings_pct + rent_pct + food_pct + transport_pct +
             utilities_pct + entertainment_pct + emergency_pct)
# assert total_pct == 1.0, "Percentages must sum to 100%!"

# ─── STEP 3: Calculate amounts ──────────────────────────────────────
savings       = monthly_salary * savings_pct
rent          = monthly_salary * rent_pct
food          = monthly_salary * food_pct
transport     = monthly_salary * transport_pct
utilities     = monthly_salary * utilities_pct
entertainment = monthly_salary * entertainment_pct
emergency     = monthly_salary * emergency_pct

# Total expenses (everything except savings)
total_expenses = (rent + food + transport + utilities +
                  entertainment + emergency)

# ─── STEP 4: Display the budget report ─────────────────────────────
print()
print(f"  Monthly Salary  : PKR {monthly_salary:>12,.2f}")
print("─" * 50)
print(f"  {'CATEGORY':<20} {'%':>5}   {'AMOUNT (PKR)':>12}")
print("─" * 50)
print(f"  {'Savings':<20} {savings_pct*100:>5.0f}%  PKR {savings:>12,.2f}")
print(f"  {'Rent':<20} {rent_pct*100:>5.0f}%  PKR {rent:>12,.2f}")
print(f"  {'Food':<20} {food_pct*100:>5.0f}%  PKR {food:>12,.2f}")
print(f"  {'Transport':<20} {transport_pct*100:>5.0f}%  PKR {transport:>12,.2f}")
print(f"  {'Utilities':<20} {utilities_pct*100:>5.0f}%  PKR {utilities:>12,.2f}")
print(f"  {'Entertainment':<20} {entertainment_pct*100:>5.0f}%  PKR {entertainment:>12,.2f}")
print(f"  {'Emergency Fund':<20} {emergency_pct*100:>5.0f}%  PKR {emergency:>12,.2f}")
print("─" * 50)
print(f"  {'Total Expenses':<20} {(1-savings_pct)*100:>5.0f}%  PKR {total_expenses:>12,.2f}")
print(f"  {'Total':<20} {'100':>5}%  PKR {monthly_salary:>12,.2f}")
print("=" * 50)

# ─── STEP 5: Useful insights ────────────────────────────────────────
daily_budget   = total_expenses / 30
annual_savings = savings * 12

print()
print("Quick Insights:")
print(f"  Daily spending budget : PKR {daily_budget:,.2f}")
print(f"  Annual savings target : PKR {annual_savings:,.2f}")
print()
print(f"At this rate, in 5 years you'll save: PKR {annual_savings * 5:,.2f}")
print("=" * 50)
````

````python
# ── Challenge 1: Temperature Converter ─────────────────────────────────────
# Write your solution here!

# Step 1: Get Celsius temperature from user
celsius = float(input("Enter temperature in Celsius: "))

# Step 2: Apply conversion formula  F = (C × 9/5) + 32
fahrenheit = (celsius * 9/5) + 32

# Step 3: Also convert to Kelvin  K = C + 273.15
kelvin = celsius + 273.15

# Step 4: Display results
print(f"\n{celsius}°C = {fahrenheit:.2f}°F = {kelvin:.2f}K")
````

````python
# ── Challenge 2: Area Calculator ────────────────────────────────────────────
import math

radius = float(input("Enter radius of circle (cm): "))

# Area = π × r²
area = math.pi * (radius ** 2)

# Circumference = 2 × π × r
circumference = 2 * math.pi * radius

print(f"\nCircle with radius = {radius} cm")
print(f"Area          : {area:.4f} cm²")
print(f"Circumference : {circumference:.4f} cm")
````

````python
# ── Challenge 3: Grade Calculator ──────────────────────────────────────────

print("Enter marks for 5 subjects (out of 100):")
s1 = float(input("  Subject 1 (Math)     : "))
s2 = float(input("  Subject 2 (English)  : "))
s3 = float(input("  Subject 3 (Python)   : "))
s4 = float(input("  Subject 4 (Stats)    : "))
s5 = float(input("  Subject 5 (Database) : "))

# Calculate average
total   = s1 + s2 + s3 + s4 + s5
average = total / 5

# Determine grade using if-elif-else
if average >= 90:
    grade  = "A"
    remark = "Excellent! 🌟"
elif average >= 80:
    grade  = "B"
    remark = "Very Good! 👍"
elif average >= 70:
    grade  = "C"
    remark = "Good. Keep it up!"
elif average >= 60:
    grade  = "D"
    remark = "Passing. Work harder!"
else:
    grade  = "F"
    remark = "Failing. Study more! 📚"

# Display report
print("\n" + "=" * 40)
print("         📋 STUDENT REPORT CARD")
print("=" * 40)
print(f"  Math     : {s1}")
print(f"  English  : {s2}")
print(f"  Python   : {s3}")
print(f"  Stats    : {s4}")
print(f"  Database : {s5}")
print("─" * 40)
print(f"  Total    : {total}/500")
print(f"  Average  : {average:.2f}%")
print(f"  Grade    : {grade}")
print(f"  Remark   : {remark}")
print("=" * 40)
````
