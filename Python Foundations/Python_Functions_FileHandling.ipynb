## 🧠 Why Functions? The Problem They Solve

Imagine you need to calculate a grade for 500 students. Without functions:

```python
# For student 1
avg1 = (85 + 92 + 78) / 3
if avg1 >= 90: grade1 = "A"
elif avg1 >= 80: grade1 = "B"
...

# For student 2 — same code again!
avg2 = (90 + 88 + 95) / 3
if avg2 >= 90: grade2 = "A"
elif avg2 >= 80: grade2 = "B"
...
# × 500 students = 5000+ lines of repeated code 😱
```

**With a function — write once, use anywhere:**
```python
def calculate_grade(marks):      # defined ONCE
    avg = sum(marks) / len(marks)
    if avg >= 90: return "A"
    elif avg >= 80: return "B"
    ...

grade1 = calculate_grade([85, 92, 78])   # used 500 times
grade2 = calculate_grade([90, 88, 95])   # same line, different data
```

### The 3 Core Benefits of Functions:
- **DRY** — Don't Repeat Yourself (write once, use many times)
- **Readable** — code becomes self-documenting with good function names
- **Maintainable** — fix a bug in one place, fixed everywhere

# 🔧 SECTION 1: Defining Functions

## Anatomy of a Python Function

```
def   function_name  ( parameters ) :
 ↑         ↑               ↑         ↑
keyword  name you    inputs it     colon
         choose      receives      required

    """ docstring — describes what the function does """
    # function body — indented!
    result = ...
    return result    ← sends back the output
```

### Function Rules:
- Must start with `def` keyword
- Name follows same rules as variables (`snake_case`)
- Body must be **indented** (4 spaces)
- **Define first, call later** — Python reads top to bottom
- `return` sends a value back — without it, function returns `None`

### 1.1 — Your First Functions

````python
# ── Simplest Possible Function ─────────────────────────────────────────────
# No parameters, no return value — just runs a block of code

def greet():
    """Prints a greeting message."""
    print("Assalamu Alaikum!")
    print("Welcome to Python Functions.")
    print("Today we learn to write reusable code.")

# CALLING the function — use its name followed by ()
greet()           # first call
print()
greet()           # second call — same function, no copy-paste!
print()
greet()           # third call
````

````python
# ── Function with a Parameter ──────────────────────────────────────────────
# Parameter = the variable name inside the parentheses
# Argument  = the actual value you pass when calling

def greet_student(name):         # 'name' is the PARAMETER
    """Greets a student by name."""
    print(f"Assalamu Alaikum, {name}!")
    print(f"Welcome to class, {name}. Ready to learn?")

greet_student("Ahmed")           # "Ahmed" is the ARGUMENT
print()
greet_student("Fatima")          # different argument, same function
print()
greet_student("Zara")
````

````python
# ── Function with return ───────────────────────────────────────────────────
# return sends a value BACK to the caller
# The caller can store it in a variable, print it, or use it directly

def square(number):
    """Returns the square of a number."""
    result = number ** 2
    return result          # send result back

# Three ways to use the return value:

# Way 1: Store in variable
num = int(input("Enter a number to square: "))
numb = 19
answer = square(num)
print(f"7 squared = {answer}")

# Way 2: Use directly in print
print(f"9 squared = {square(9)}")

# Way 3: Use in another calculation
total = square(3) + square(4)
print(f"3² + 4²   = {total}")

print()

# Without return — function returns None
def no_return(x):
    """Calculates but does not return."""
    result = x * 2
    # no return statement!
    return result

value = no_return(5)
print(f"no_return(5) gave us: {value}")   # None!
````

### 1.2 — Docstrings: Documenting Your Functions

````python
# ── Docstrings ─────────────────────────────────────────────────────────────
# A docstring is a string at the very start of a function
# It describes: what the function does, parameters, and return value
# Professional code ALWAYS has docstrings

def calculate_bmi(weight_kg, height_m=1.75):
    """
    Calculate Body Mass Index (BMI).

    Parameters:
        weight_kg (float): Weight in kilograms
        height_m  (float): Height in meters

    Returns:
        float: BMI value rounded to 2 decimal places

    Example:
        >>> calculate_bmi(70, 1.75)
        22.86
    """
    bmi = weight_kg / (height_m ** 2)
    return round(bmi, 2)

# Call the function
bmi = calculate_bmi(80, height_m= 1.80)
print(f"BMI: {bmi}")

# Read the docstring using help()
print()
help(calculate_bmi)
````

# ⚙️ SECTION 2: Parameters & Return Values

## Types of Parameters

Python functions support several types of parameters — each solves a different problem:

| Type | Syntax | Description |
|------|--------|-------------|
| **Positional** | `def f(a, b)` | Must be passed in order |
| **Default** | `def f(a, b=10)` | Optional — has a fallback value |
| **Keyword** | `f(b=5, a=2)` | Pass by name — order doesn't matter |
| **`*args`** | `def f(*args)` | Accept any number of positional arguments |
| **`**kwargs`** | `def f(**kwargs)` | Accept any number of keyword arguments |

### 2.1 — Positional and Default Parameters

````python
# ── Positional Parameters ─────────────────────────────────────────────────
# Passed in ORDER — position matters!

def introduce(name, age, city):
    """Introduces a person."""
    print(f"My name is {name}, I am {age} years old, from {city}.")

introduce("Ahmed", 22, "Islamabad")    # correct order
introduce("Fatima", 20, "Lahore")

# Order matters!
introduce(22, "Ahmed", "Islamabad")    # WRONG order — name = 22, age = "Ahmed"!

print()

# ── Default Parameters ─────────────────────────────────────────────────────
# Give a parameter a default value
# If caller doesn't provide it → default is used
# Default parameters MUST come AFTER non-default ones

def create_account(name, account_type="Savings", currency="PKR", active=True):
    """
    Creates a bank account.
    Only 'name' is required — rest have defaults.
    """
    status = "Active" if active else "Inactive"
    print(f"Account: {name} | Type: {account_type} | Currency: {currency} | {status}")

# All defaults used
create_account("Ahmed")

# Override one default
create_account("Fatima", currency="USD")

# Override multiple defaults
create_account("Zara", "Savings", "USD")

# Override all
create_account("Hassan", "Business", "EUR", False)
````

````python
# ── Keyword Arguments ─────────────────────────────────────────────────────
# Pass arguments by NAME — order no longer matters!
# Makes function calls more readable

def send_email(to, subject, body, cc="", priority="Normal"):
    """Simulates sending an email."""
    print(f"To      : {to}")
    print(f"CC      : {cc if cc else 'None'}")
    print(f"Subject : {subject}")
    print(f"Priority: {priority}")
    print(f"Body    : {body[:50]}...")
    print()

# Using keyword arguments — any order!
send_email(
    to       = "manager@company.com",
    subject  = "Monthly Report",
    body     = "Please find the monthly sales analysis attached.",
    priority = "High"
)

# Mix of positional and keyword (positional MUST come first)
send_email(
    "team@company.com",           # positional
    "Meeting Tomorrow",           # positional
    body     = "We have a team meeting at 10am.",
    cc       = "hr@company.com"
)
````

### 2.2 — `*args` and `**kwargs`

````python
# ── *args — Accept ANY number of positional arguments ─────────────────────
# *args collects all extra positional arguments into a TUPLE
# The name 'args' is convention — the * is what matters

def calculate_total(*prices):
    """
    Calculates total of any number of prices.
    *prices collects all arguments into a tuple.
    """
    print(f"Prices received: {prices}")
    print(f"Type           : {type(prices)}")
    total = sum(prices)
    return total

# Call with any number of arguments!
print(f"Total (2 items) : PKR {calculate_total(1200, 850)}")
print()
print(f"Total (5 items) : PKR {calculate_total(1200, 850, 3500, 200, 4800):,}")
print()

# Real use: flexible statistics function
def summarize(*numbers):
    """Prints summary stats for any number of values."""
    print(f"Count   : {len(numbers)}")
    print(f"Sum     : {sum(numbers)}")
    print(f"Average : {sum(numbers)/len(numbers):.2f}")
    print(f"Min     : {min(numbers)}")
    print(f"Max     : {max(numbers)}")

print("=== Summary ===")
summarize(85, 92, 78, 95, 88, 73, 61)
````

````python
# ── **kwargs — Accept ANY number of keyword arguments ─────────────────────
# **kwargs collects all extra keyword arguments into a DICTIONARY
# Very useful when you don't know in advance what keys will be provided

def create_student_profile(**details):
    """
    Creates a student profile from any keyword arguments.
    **details collects all keyword args into a dict.
    """
    print(f"Received: {details}")
    print(f"Type    : {type(details)}")
    print()
    print("=== Student Profile ===")
    for key, value in details.items():
        print(f"  {key.replace('_', ' ').title():<18}: {value}")

create_student_profile(
    name        = "Ahmed Khan",
    age         = 22,
    city        = "Islamabad",
    department  = "Computer Science",
    gpa         = 3.85,
    semester    = 6,
    scholarship = True
)
````

````python
# ── Combining all parameter types ─────────────────────────────────────────
# Order must be: positional → *args → keyword defaults → **kwargs

def full_report(title, *data_points, separator="─", width=40, **metadata):
    """
    Prints a formatted report.
    title      : required positional
    *data_points : any number of data values
    separator  : default keyword arg
    width      : default keyword arg
    **metadata : any extra keyword info
    """
    print(separator * width)
    print(f"  {title.upper()}")
    print(separator * width)

    for i, point in enumerate(data_points, 1):
        print(f"  [{i}] {point}")

    if metadata:
        print(separator * width)
        for key, val in metadata.items():
            print(f"  {key}: {val}")
    print(separator * width)

full_report(
    "Monthly Sales Summary",
    "January: PKR 82,000",
    "February: PKR 95,000",
    "March: PKR 110,000",
    separator = "=",
    width     = 38,
    author    = "Ahmed Khan",
    date      = "2024-03-15",
    version   = "v1.2"
)
````

### 2.3 — Return Values

````python
# ── Returning Multiple Values ──────────────────────────────────────────────
# Python functions can return MULTIPLE values
# They come back as a TUPLE — unpack them!

def analyze_scores(scores):
    """
    Analyzes a list of scores.

    Returns:
        tuple: (minimum, maximum, average, count)
    """
    minimum = min(scores)
    maximum = max(scores)
    average = sum(scores) / len(scores)
    count   = len(scores)
    return minimum, maximum, average, count   # returns a tuple

marks = [85, 92, 78, 95, 88, 73, 61, 90]

# Option 1: Unpack into separate variables
low, high, avg, n = analyze_scores(marks)
print(f"Count  : {n}")
print(f"Min    : {low}")
print(f"Max    : {high}")
print(f"Average: {avg:.2f}")

print()

# Option 2: Capture as a tuple
result = analyze_scores(marks)
print(f"Result tuple: {result}")
print(f"Access max  : {result[1]}")
````

````python
# ── Early Return ───────────────────────────────────────────────────────────
# return can appear anywhere — the function stops immediately
# Useful for input validation and guard clauses

def safe_divide(numerator, denominator):
    """
    Divides two numbers safely.
    Returns None and prints an error if denominator is 0.
    """
    # Guard clause — return early if input is invalid
    if denominator == 0:
        print("❌ Error: Cannot divide by zero!")
        return None    # exit immediately

    if not isinstance(numerator, (int, float)) or not isinstance(denominator, (int, float)):
        print("❌ Error: Both inputs must be numbers!")
        return None

    # Only reaches here if inputs are valid
    result = numerator / denominator
    return round(result, 4)

print(safe_divide(10, 2))      # valid
print(safe_divide(10, 0))      # division by zero
print(safe_divide(15, 4))      # valid
print(safe_divide("a", 2))    # invalid type
````

````python
# ── Variable Scope — Local vs Global ──────────────────────────────────────
# Variables created INSIDE a function are LOCAL — invisible outside
# Variables created OUTSIDE functions are GLOBAL — visible everywhere

company_name = "TechSoft"     # GLOBAL variable

def calculate_bonus(salary):
    """Calculates annual bonus."""
    bonus_rate = 0.15          # LOCAL variable — only exists inside this function
    bonus      = salary * bonus_rate
    print(f"  Company     : {company_name}")   # can access GLOBAL
    print(f"  Salary      : PKR {salary:,}")
    print(f"  Bonus rate  : {bonus_rate*100:.0f}%")
    print(f"  Bonus amount: PKR {bonus:,.0f}")
    return bonus

calculate_bonus(80000)

print()

# bonus_rate does NOT exist outside the function
try:
    print(bonus_rate)
except NameError as e:
    print(f"❌ NameError: {e}")
    print("   'bonus_rate' is LOCAL — it only lives inside calculate_bonus()")
````

### 2.4 — Lambda Functions (Anonymous Functions)

````python
# ── Lambda Functions ───────────────────────────────────────────────────────
# A lambda is a tiny, one-line function without a name
# Syntax: lambda parameters: expression
# Used for short, simple operations — especially with sort(), filter(), map()

# Regular function
def double(x):
    return x * 2

# Equivalent lambda
double_lambda = lambda x: x * 2

print(f"Regular function: {double(7)}")
print(f"Lambda function : {double_lambda(7)}")

print()

# Lambda with multiple parameters
add      = lambda a, b: a + b
power    = lambda base, exp: base ** exp
greeting = lambda name, city: f"Hello {name} from {city}!"

print(f"add(3, 5)           = {add(3, 5)}")
print(f"power(2, 8)         = {power(2, 8)}")
print(f"greeting(Ahmed, ISB)= {greeting('Ahmed', 'Islamabad')}")

print()

# Most common use: sorting with a custom key
students = [
    {"name": "Zara",   "gpa": 3.71},
    {"name": "Ahmed",  "gpa": 3.85},
    {"name": "Fatima", "gpa": 3.92},
    {"name": "Hassan", "gpa": 3.55},
]

# Sort by GPA (highest first) using lambda
sorted_students = sorted(students, key=lambda s: s["gpa"], reverse=True)
print("Ranked by GPA:")
for rank, s in enumerate(sorted_students, 1):
    print(f"  {rank}. {s['name']:<10} GPA: {s['gpa']}")
````

### 2.5 — Built-in Functional Tools: `map()`, `filter()`

````python
# ── map() — Apply a function to every item in a list ─────────────────────
# map(function, iterable) → returns a map object (convert to list)

prices = [1000, 2500, 800, 4200, 1500]

# Apply 15% tax to every price using map
def add_tax(price):
    return price * 1.15

prices_with_tax = list(map(add_tax, prices))
print(f"Original  : {prices}")
print(f"With tax  : {prices_with_tax}")

# Shorter: use lambda with map
prices_with_tax = list(map(lambda p: round(p * 1.15, 2), prices))
print(f"With tax  : {prices_with_tax}")

print()

# ── filter() — Keep only items that pass a condition ──────────────────────
# filter(function, iterable) → keeps items where function returns True

scores = [85, 42, 96, 73, 58, 91, 67, 38, 78, 55]

# Keep only passing scores (>= 60)
passing = list(filter(lambda s: s >= 60, scores))
failing = list(filter(lambda s: s < 60,  scores))

print(f"All scores: {scores}")
print(f"Passing   : {passing}")
print(f"Failing   : {failing}")

# Filter names longer than 5 characters
names = ["Ahmed", "Fatima", "Zara", "Hassan", "Sara", "Muhammad"]
long_names = list(filter(lambda n: len(n) > 5, names))
print(f"\nNames > 5 chars: {long_names}")
````

### 2.6 — Building Reusable Data Analysis Functions

````python
# ── A Reusable Function Library for Data Analysis ─────────────────────────
# These are the kinds of functions you'll write in real projects

def clean_name(name):
    """Strips whitespace and converts to Title Case."""
    return name.strip().title()

def calculate_average(values):
    """Returns average of a list. Returns None if list is empty."""
    if not values:
        return None
    return round(sum(values) / len(values), 2)

def assign_grade(average):
    """Returns letter grade based on average score."""
    if average is None: return "N/A"
    if average >= 90:   return "A"
    elif average >= 80: return "B"
    elif average >= 70: return "C"
    elif average >= 60: return "D"
    else:               return "F"

def format_currency(amount, symbol="PKR"):
    """Formats a number as currency string."""
    return f"{symbol} {amount:,.2f}"

def percentage_change(old_value, new_value):
    """Calculates % change between two values."""
    if old_value == 0:
        return None
    change = ((new_value - old_value) / abs(old_value)) * 100
    return round(change, 2)

def normalize(values):
    """Normalizes a list of numbers to 0-1 range."""
    min_v = min(values)
    max_v = max(values)
    if max_v == min_v:
        return [0] * len(values)
    return [round((v - min_v) / (max_v - min_v), 4) for v in values]

# ── Test all functions ─────────────────────────────────────────────────────
raw_names  = ["  ahmed ", "FATIMA", "zara  "]
marks      = [85, 92, 78, 95, 88]
jan_sales  = 82000
feb_sales  = 95000

print("=== Testing Our Function Library ===")
print()

print("clean_name():")
for n in raw_names:
    print(f"  '{n}' → '{clean_name(n)}'")

avg = calculate_average(marks)
print(f"\ncalculate_average({marks}): {avg}")
print(f"assign_grade({avg})       : {assign_grade(avg)}")
print(f"format_currency(85000)   : {format_currency(85000)}")
print(f"percentage_change(82→95K): {percentage_change(jan_sales, feb_sales)}%")
print(f"normalize({marks})       :")
print(f"  {normalize(marks)}")
````


### 3.1 — Reading CSV with the Built-in `csv` Module

````python
# ── Reading CSV with csv.reader ────────────────────────────────────────────
# csv.reader reads each row as a LIST of strings

import csv

print("=== Reading students.csv with csv.reader ===")
print()

with open("students.csv", "r", encoding="utf-8") as file:
    reader = csv.reader(file)

    # Read header row
    header = next(reader)       # next() reads ONE row
    print(f"Header columns: {header}")
    print(f"Column count  : {len(header)}")
    print()

    # Read data rows
    print(f"{'NAME':<10} {'AGE':>5} {'CITY':<12} {'MATH':>6}")
    print("-" * 36)
    for row in reader:
        name, age, city, math = row[0], row[1], row[2], row[3]
        print(f"{name:<10} {age:>5} {city:<12} {math:>6}")

# ⚠️ Important: All values from csv.reader are STRINGS!
# Need int() or float() to do math
````

````python
# ── Reading CSV with csv.DictReader ───────────────────────────────────────
# DictReader reads each row as a DICTIONARY — much more readable!
# Keys are automatically the header column names

import csv

print("=== Reading with csv.DictReader ===")
print()

students = []   # we'll store all rows here

with open("students.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)

    for row in reader:
        # Each row is a dict like:
        # {'name':'Ahmed', 'age':'22', 'city':'Islamabad', ...}

        # Convert numeric strings to numbers
        student = {
            "name"   : row["name"],
            "age"    : int(row["age"]),
            "city"   : row["city"],
            "math"   : int(row["math"]),
            "english": int(row["english"]),
            "python" : int(row["python"]),
            "stats"  : int(row["stats"]),
        }
        students.append(student)

print(f"Records loaded: {len(students)}")
print(f"\nFirst record  : {students[0]}")
print(f"Last record   : {students[-1]}")
print()

# Now we can DO things with the data
print("=== Students from Islamabad ===")
isb_students = [s for s in students if s["city"] == "Islamabad"]
for s in isb_students:
    print(f"  {s['name']} (age {s['age']})")
````

# 💾 SECTION 4: Writing to Files

## File Modes — How You Open a File

When opening a file with `open()`, you specify a **mode**:

| Mode | Symbol | Description |
|------|--------|-------------|
| Read | `"r"` | Read only. Error if file doesn't exist. |
| Write | `"w"` | Write. **Creates new** or **overwrites** existing! |
| Append | `"a"` | Add to end. Creates if doesn't exist. |
| Read+Write | `"r+"` | Both read and write. File must exist. |

## The `with` Statement — Always Use This!

```python
# ❌ Risky way — file might not close if error occurs
f = open("file.txt", "w")
f.write("data")
f.close()    # you must manually close!

# ✅ Safe way — file closes automatically, even if error occurs
with open("file.txt", "w") as f:
    f.write("data")
# file is automatically closed here
```

> 🏆 **Always use `with open(...)` — it's safer, cleaner, and the professional standard.**

### 4.1 — Writing Text Files

````python
# ── Writing to a Text File ─────────────────────────────────────────────────

# Write mode 'w' — creates new or OVERWRITES existing
with open("report.txt", "w", encoding="utf-8") as f:
    f.write("STUDENT PERFORMANCE REPORT\n")
    f.write("=" * 40 + "\n")
    f.write("Generated: 2024-03-15\n")
    f.write("=" * 40 + "\n\n")

print("✅ report.txt created")

# Append mode 'a' — adds to existing file without erasing
students_summary = [
    ("Ahmed",   85.75, "B"),
    ("Fatima",  88.75, "B"),
    ("Zara",    72.50, "C"),
    ("Hassan",  61.25, "D"),
    ("Sara",    95.00, "A"),
]

with open("report.txt", "a", encoding="utf-8") as f:
    f.write(f"{'NAME':<12} {'AVERAGE':>8} {'GRADE':>6}\n")
    f.write("-" * 30 + "\n")
    for name, avg, grade in students_summary:
        f.write(f"{name:<12} {avg:>8.2f} {grade:>6}\n")
    f.write("-" * 30 + "\n")
    f.write(f"\nTotal students: {len(students_summary)}\n")

print("✅ Summary appended to report.txt")
print()

# Read it back to verify
print("=== Contents of report.txt ===")
with open("report.txt", "r", encoding="utf-8") as f:
    print(f.read())
````

## Scenario:
You are a **junior data analyst** at a retail company. Your manager sends you a raw sales CSV and asks you to:

2. Clean and enrich it (calculate revenue, profit)

This is a **real-world workflow** — exactly what you will do on the job.

````python
# ── STEP 1: Create raw sales data ─────────────────────────────────────────

import csv, os

raw_sales = [
    ["date",       "product",   "category",     "units", "unit_price", "unit_cost", "region",  "salesperson"],
    ["2024-01-05", "Laptop",    "Electronics",  12,      85000,        60000,       "North",   "Ahmed"],
    ["2024-01-08", "Mouse",     "Accessories",  85,      1500,         800,         "South",   "Fatima"],
    ["2024-01-12", "Keyboard",  "Accessories",  45,      3500,         1800,        "East",    "Zara"],
    ["2024-01-15", "Monitor",   "Electronics",  8,       42000,        28000,       "North",   "Ahmed"],
    ["2024-01-20", "Laptop",    "Electronics",  18,      85000,        60000,       "West",    "Hassan"],
    ["2024-01-22", "USB Hub",   "Accessories",  120,     2000,         900,         "South",   "Fatima"],
    ["2024-02-03", "Laptop",    "Electronics",  22,      85000,        60000,       "East",    "Zara"],
    ["2024-02-07", "Headphones","Electronics",  35,      8500,         4200,        "North",   "Ahmed"],
    ["2024-02-14", "Mouse",     "Accessories",  100,     1500,         800,         "West",    "Hassan"],
    ["2024-02-19", "Laptop",    "Electronics",  15,      85000,        60000,       "South",   "Fatima"],
    ["2024-02-25", "Monitor",   "Electronics",  12,      42000,        28000,       "East",    "Zara"],
    ["2024-03-02", "Keyboard",  "Accessories",  60,      3500,         1800,        "North",   "Ahmed"],
    ["2024-03-10", "Laptop",    "Electronics",  28,      85000,        60000,       "West",    "Hassan"],
    ["2024-03-15", "USB Hub",   "Accessories",  150,     2000,         900,         "South",   "Fatima"],
    ["2024-03-22", "Headphones","Electronics",  42,      8500,         4200,        "North",   "Ahmed"],
]

with open("raw_sales.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerows(raw_sales)

print(f"✅ raw_sales.csv created with {len(raw_sales)-1} records")
````

````python
# ── STEP 2: Define analysis functions ─────────────────────────────────────

def load_sales(filepath):
    """
    Loads and enriches sales data from CSV.
    Adds: revenue, cost_total, profit, profit_margin, month
    """
    sales = []
    with open(filepath, "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            units      = int(row["units"])
            unit_price = int(row["unit_price"])
            unit_cost  = int(row["unit_cost"])

            revenue       = units * unit_price
            cost_total    = units * unit_cost
            profit        = revenue - cost_total
            margin        = round((profit / revenue) * 100, 2)
            month         = row["date"][:7]   # "2024-01-05" → "2024-01"

            sales.append({
                "date"         : row["date"],
                "month"        : month,
                "product"      : row["product"],
                "category"     : row["category"],
                "units"        : units,
                "unit_price"   : unit_price,
                "unit_cost"    : unit_cost,
                "revenue"      : revenue,
                "cost_total"   : cost_total,
                "profit"       : profit,
                "margin_pct"   : margin,
                "region"       : row["region"],
                "salesperson"  : row["salesperson"],
            })
    return sales

def group_by(data, key, value_key, agg="sum"):
    """
    Groups data by a key and aggregates a value field.
    agg: 'sum', 'avg', 'count', 'max', 'min'
    """
    groups = {}
    for row in data:
        group_val = row[key]
        if group_val not in groups:
            groups[group_val] = []
        groups[group_val].append(row[value_key])

    result = {}
    for group, values in groups.items():
        if   agg == "sum":   result[group] = sum(values)
        elif agg == "avg":   result[group] = round(sum(values)/len(values), 2)
        elif agg == "count": result[group] = len(values)
        elif agg == "max":   result[group] = max(values)
        elif agg == "min":   result[group] = min(values)

    return dict(sorted(result.items(), key=lambda x: x[1], reverse=True))

def print_table(title, data_dict, key_label="Item", val_label="Value",
                prefix="PKR ", width=44):
    """Prints a formatted two-column table."""
    print("=" * width)
    print(f"  {title}")
    print("=" * width)
    print(f"  {key_label:<18} {val_label:>20}")
    print("  " + "-" * (width - 2))
    for key, val in data_dict.items():
        if isinstance(val, float):
            val_str = f"{prefix}{val:>,.2f}"
        else:
            val_str = f"{prefix}{val:>,.0f}"
        print(f"  {str(key):<18} {val_str:>20}")
    print("=" * width)

print("✅ Functions defined: load_sales(), group_by(), print_table()")
````

````python
# ── STEP 3: Load and inspect ───────────────────────────────────────────────

sales = load_sales("raw_sales.csv")

print(f"Records loaded: {len(sales)}")
print(f"\nFirst record  : ")
for k, v in sales[0].items():
    print(f"  {k:<15}: {v}")

print()

# Quick totals
total_revenue = sum(s["revenue"] for s in sales)
total_profit  = sum(s["profit"]  for s in sales)
total_units   = sum(s["units"]   for s in sales)
avg_margin    = sum(s["margin_pct"] for s in sales) / len(sales)

print("=" * 44)
print("  📊 OVERALL SUMMARY")
print("=" * 44)
print(f"  Total Records  : {len(sales)}")
print(f"  Total Units    : {total_units:,}")
print(f"  Total Revenue  : PKR {total_revenue:>12,}")
print(f"  Total Profit   : PKR {total_profit:>12,}")
print(f"  Avg Margin     : {avg_margin:.1f}%")
print("=" * 44)
````

````python
# ── STEP 4: Multi-dimensional analysis ────────────────────────────────────

# Revenue by product
by_product = group_by(sales, "product", "revenue", "sum")
print_table("💻 Revenue by Product", by_product)

print()

# Profit by region
by_region = group_by(sales, "region", "profit", "sum")
print_table("🗺️  Profit by Region", by_region)

print()

# Revenue by salesperson
by_person = group_by(sales, "salesperson", "revenue", "sum")
print_table("🏆 Revenue by Salesperson", by_person)

print()

# Revenue by month
by_month = group_by(sales, "month", "revenue", "sum")
print_table("📅 Revenue by Month", by_month)

print()

# Best margin by product
by_margin = group_by(sales, "product", "margin_pct", "avg")
print_table("📈 Avg Margin by Product", by_margin, val_label="Margin %", prefix="%  ")
````

````python
# ── STEP 5: Save all results to files ─────────────────────────────────────

import csv

# 1. Save enriched sales data to CSV
fields = ["date", "month", "product", "category", "units",
          "unit_price", "revenue", "cost_total", "profit",
          "margin_pct", "region", "salesperson"]

with open("sales_enriched.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fields, extrasaction="ignore")
    writer.writeheader()
    writer.writerows(sales)

print("✅ sales_enriched.csv saved")

# 2. Save text summary report
with open("sales_summary.txt", "w", encoding="utf-8") as f:
    f.write("SALES ANALYSIS REPORT\n")
    f.write("=" * 44 + "\n")
    f.write(f"Total Records  : {len(sales)}\n")
    f.write(f"Total Revenue  : PKR {total_revenue:,}\n")
    f.write(f"Total Profit   : PKR {total_profit:,}\n")
    f.write(f"Avg Margin     : {avg_margin:.1f}%\n\n")

    f.write("TOP PRODUCTS BY REVENUE:\n")
    f.write("-" * 44 + "\n")
    for product, rev in by_product.items():
        f.write(f"  {product:<15}: PKR {rev:>12,}\n")

    f.write("\nTOP SALESPERSONS:\n")
    f.write("-" * 44 + "\n")
    for person, rev in by_person.items():
        f.write(f"  {person:<15}: PKR {rev:>12,}\n")

print("✅ sales_summary.txt saved")
print()

# Read summary back
with open("sales_summary.txt", "r") as f:
    print(f.read())

# Final file list
print("=== All output files ===")
import os
for file in sorted(os.listdir(".")):
    if file.endswith((".csv", ".txt")):
        size = os.path.getsize(file)
        print(f"  📄 {file:<35} {size:>6} bytes")
````

| Concept | Syntax | Remember |
