## 🧠 What Is Control Flow?

By default, Python runs code **line by line, top to bottom**. But real programs need to:
- Make **decisions** → `if / elif / else`
- **Repeat** actions → `for` and `while` loops
- Store **structured data** → Dictionaries

**Without control flow:**
```
Line 1 → Line 2 → Line 3 → Line 4 → Done.
```

**With control flow:**
```
Line 1 → Decision? → Yes: Line 2 → Repeat 10 times → Line 3 → Done.
                   → No:  Line 4 → Done.
```

Control flow is what makes programs **intelligent and dynamic**.

# 📚 SECTION 1: Dictionaries

## What Is a Dictionary?

A **dictionary** stores data as **key: value pairs** — like a real dictionary where you look up a **word (key)** to find its **meaning (value)**.

```
Real dictionary:        Python dictionary:
"apple" → a fruit       "name"  → "Ahmed"
"run"   → to move fast  "age"   → 22
"blue"  → a color       "city"  → "Islamabad"
```

### Why Dictionaries?

With a **list**, you access by position: `student[0]` — you must remember index 0 = name, index 1 = age...

With a **dictionary**, you access by name: `student["name"]` — much more readable!

### Key Rules:
- Keys must be **unique** — 
- Keys must be **immutable** — 
- Values can be **anything** — 
- Dictionaries are **ordered** -
- Dictionaries are **mutable** —

### 1.1 — Creating Dictionaries

````python
# ── Creating Dictionaries ──────────────────────────────────────────────────
# Syntax: {key: value, key: value, ...}
# Use curly braces {}, keys and values separated by colon :

# Simple student dictionary
student = {
    "name"      : "Ahmed",
    "age"       : 22,
    "city"      : "Islamabad",
    "gpa"       : 3.85,
    "is_active" : True
}

# Product dictionary
product = {
    "id"       : 1001,
    "name"     : "Laptop",
    "price"    : 85000,
    "in_stock" : True,
    "brand"    : "Dell"
}

# Empty dictionary (add items later)
empty_dict = {}

# Dictionary with a list as a value!
student_marks = {
    "name"    : "Fatima",
    "marks"   : [85, 92, 78, 95, 88],   # list inside dict
    "subjects": ["Math", "English", "Python", "Stats", "Database"]
}

print("Student  :", student)
print("Product  :", product)
print("Empty    :", empty_dict)
print("Marks    :", student_marks)
print()
print("Type     :", type(student))
print("Keys     :", len(student), "items")
````

### 1.2 — Accessing Values

````python
# ── Accessing Values ───────────────────────────────────────────────────────
# Method 1: Square bracket notation dict["key"]
# Method 2: .get("key") — safer, no error if key missing

student = {
    "name" : "Ahmed",
    "age"  : 22,
    "city" : "Islamabad",
    "gpa"  : 3.85
}

# Method 1: Square brackets
print("=== Square Bracket Access ===")
print("Name :", student["name"])
print("Age  :", student["age"])
print("GPA  :", student["gpa"])

print()

# Method 2: .get() — returns None (or a default) if key doesn't exist
print("=== .get() Access ===")
print("City    :", student.get("city"))
print("Phone   :", student.get("phone"))                    # key missing → None
print("Phone   :", student.get("phone", "Not provided"))    # custom default value
print("Country :", student.get("country", "Pakistan"))      # custom default

print()

# Square bracket vs .get() — the key difference:
# student["phone"]      → KeyError (crashes!)
# student.get("phone")  → None (safe)
print("=== Demonstrating KeyError ===")
try:
    print(student["phone"])   # This key doesn't exist
except KeyError as e:
    print(f"❌ KeyError: key {e} does not exist in the dictionary")
    print("   → Use .get() to avoid this error!")
````

### 1.3 — Adding, Updating & Removing Items

````python
# ── Adding and Updating Items ──────────────────────────────────────────────
# Both use the same syntax: dict["key"] = value
# If key EXISTS  → updates the value
# If key MISSING → creates a new key-value pair

student = {"name": "Ahmed", "age": 22, "city": "Islamabad"}
print("Original:", student)

# Add a new key
student["gpa"]   = 3.85
student["phone"] = "0300-1234567"
print("After adding gpa & phone:", student)

# Update an existing key
student["age"]  = 23           # birthday!
student["city"] = "Lahore"     # moved city
print("After updating age & city:", student)

# Update multiple keys at once using .update()
student.update({
    "gpa"     : 3.92,
    "semester": 6,
    "dept"    : "Computer Science"
})
print("After .update():", student)
````

````python
# ── Removing Items ─────────────────────────────────────────────────────────
# del      → removes a key, no return value, error if missing
# .pop()   → removes and RETURNS the value, error if missing (unless default given)
# .clear() → removes ALL items

student = {
    "name" : "Ahmed",
    "age"  : 22,
    "city" : "Islamabad",
    "gpa"  : 3.85,
    "phone": "0300-1234567"
}
print("Original:", student)

# del — remove a key permanently
del student["phone"]
print("After del phone:", student)

# .pop() — remove AND get the value back
removed_gpa = student.pop("gpa")
print(f"Popped GPA value: {removed_gpa}")
print("After pop gpa   :", student)

# .pop() with a default (safe — no error if key missing)
result = student.pop("email", "No email found")
print(f"Pop missing key : {result}")

# .popitem() — removes and returns the LAST inserted key-value pair
last_item = student.popitem()
print(f"popitem()       : {last_item}")
print("After popitem() :", student)
````

### 1.4 — Looping Through Dictionaries

````python
# ── Dictionary Methods: keys(), values(), items() ──────────────────────────
# These three are the most important dictionary methods

student = {
    "name" : "Ahmed",
    "age"  : 22,
    "city" : "Islamabad",
    "gpa"  : 3.85,
    "dept" : "Computer Science"
}

# .keys() — all the keys
print("Keys   :", list(student.keys()))

# .values() — all the values
print("Values :", list(student.values()))

# .items() — all key-value PAIRS (as tuples)
print("Items  :", list(student.items()))

print()

# Loop through keys
print("=== Looping through keys ===")
for key in student:
    print(f"  {key}")

print()

# Loop through keys AND values together (most common!)
print("=== Looping through items ===")
for key, value in student.items():
    print(f"  {key:<10} : {value}")
````

````python
# ── Nested Dictionaries ────────────────────────────────────────────────────
# A dictionary where values are also dictionaries
# Very common in real data (JSON format, API responses)

company = {
    "name": "TechSoft Pakistan",
    "employees": {
        "E001": {"name": "Ahmed",  "role": "Data Analyst", "salary": 85000},
        "E002": {"name": "Fatima", "role": "Python Dev",   "salary": 95000},
        "E003": {"name": "Zara",   "role": "ML Engineer",  "salary": 120000},
    },
    "location": "Islamabad"
}

# Access nested values — chain the keys
print("Company :", company["name"])
print("Location:", company["location"])
print()

# Access E001's details
print("Employee E001 :", company["employees"]["E001"])
print("E001 Name     :", company["employees"]["E001"]["name"])
print("E001 Salary   :", company["employees"]["E001"]["salary"])

print()

# Loop through nested dict
print("=== Employee Directory ===")
print(f"  {'ID':<6} {'NAME':<10} {'ROLE':<16} {'SALARY':>10}")
print("  " + "-" * 46)
for emp_id, info in company["employees"].items():
    print(f"  {emp_id:<6} {info['name']:<10} {info['role']:<16} PKR {info['salary']:>7,}")

# Total salary bill
total_salary = sum(info["salary"] for info in company["employees"].values())
print(f"  {'':>32} {'─'*14}")
print(f"  {'Total Salary Bill':>32}  PKR {total_salary:>7,}")
````

# 🔀 SECTION 2: `if`, `elif`, `else` — Making Decisions

## What Is a Conditional Statement?

A conditional lets your program **choose different paths** based on whether something is `True` or `False`.

**Real life analogy:**
```
IF it is raining:
    take an umbrella
ELIF it is cloudy:
    wear a jacket
ELSE:
    wear sunglasses
```

**Python syntax:**
```python
if condition:          # colon required!
    # code here        # MUST be indented (4 spaces or 1 tab)
elif other_condition:  # optional, as many as needed
    # code here
else:                  # optional, runs if nothing above was True
    # code here
```

## ⚠️ Indentation Is NOT Optional in Python!

Python uses **indentation (spaces/tabs)** to know what code belongs inside the `if`.  
This is different from most languages that use `{}` curly braces.

```python
if True:
    print("inside if")     # ← 4 spaces indent — INSIDE the if
print("outside if")        # ← no indent — OUTSIDE the if, always runs
```

### 2.1 — Comparison & Logical Operators

````python
# ── Comparison Operators ───────────────────────────────────────────────────
# These return True or False and are used in conditions

x = 10
y = 20

print(f"x = {x}, y = {y}")
print("─" * 35)
print(f"x == y   (equal)           : {x == y}")
print(f"x != y   (not equal)       : {x != y}")
print(f"x >  y   (greater than)    : {x >  y}")
print(f"x <  y   (less than)       : {x <  y}")
print(f"x >= y   (greater or equal): {x >= y}")
print(f"x <= y   (less or equal)   : {x <= y}")

print()

# ⚠️ Common beginner mistake:
# = is ASSIGNMENT (store value)
# == is COMPARISON (check if equal)
age = 22                # stores 22 in age
print(age == 22)        # checks if age equals 22 → True
print(age == 18)        # checks if age equals 18 → False
````

````python
# ── Logical Operators ──────────────────────────────────────────────────────
# Combine multiple conditions together
# and → BOTH must be True
# or  → AT LEAST ONE must be True
# not → REVERSES the boolean

age        = 22
has_id     = True
has_ticket = False
score      = 75

print("=== and — BOTH must be True ===")
print(f"age > 18 and has_id    : {age > 18 and has_id}")
print(f"age > 18 and has_ticket: {age > 18 and has_ticket}")

print()

print("=== or — AT LEAST ONE True ===")
print(f"has_id or has_ticket   : {has_id or has_ticket}")
print(f"age < 18 or score > 90 : {age < 18 or score > 90}")
print(f"age > 18 or score > 60 : {age > 18 or score > 60}")

print()

print("=== not — REVERSES boolean ===")
print(f"has_ticket             : {has_ticket}")
print(f"not has_ticket         : {not has_ticket}")
print(f"not (age > 18)         : {not (age > 18)}")

print()

# Chaining conditions
print("=== Chained Conditions ===")
print(f"18 < age < 30          : {18 < age < 30}")   # Python allows this!
print(f"60 <= score <= 80      : {60 <= score <= 80}")
````

### 2.2 — if, elif, else

````python
# ── Simple if / else ──────────────────────────────────────────────────────

age = 20

# Simple if — only runs if condition is True
if age >= 18:
    print("You are an adult.")
    print("You can vote in elections.")

# if / else — one of two paths
temperature = 12   # degrees Celsius

if temperature > 30:
    print("It's hot! Wear light clothes.")
else:
    print("It's cool. Bring a jacket.")

print()

# Grade check example
score = 78
if score >= 60:
    status = "PASSED ✅"
else:
    status = "FAILED ❌"
print(f"Score: {score} → {status}")
````

````python
# ── if / elif / else — Multiple Conditions ────────────────────────────────
# Python checks each condition TOP to BOTTOM
# As soon as ONE is True, it runs that block and SKIPS the rest

score = 83

if score >= 90:
    grade  = "A"
    remark = "Outstanding! 🌟"
elif score >= 80:
    grade  = "B"
    remark = "Very Good! 👍"
elif score >= 70:
    grade  = "C"
    remark = "Good. Keep it up!"
elif score >= 60:
    grade  = "D"
    remark = "Passing. Study harder!"
else:
    grade  = "F"
    remark = "Failed. Please retake. 📚"

print(f"Score  : {score}")
print(f"Grade  : {grade}")
print(f"Remark : {remark}")
````

````python
# ── Nested if statements ───────────────────────────────────────────────────
# An if inside another if
# Use carefully — too many levels makes code hard to read

age       = 25
has_cnic  = True
has_money = True
amount    = 5000

print("=== Bank Loan Eligibility Check ===")

if age >= 21:
    print(f"✅ Age {age}: Meets minimum age requirement")

    if has_cnic:
        print("✅ CNIC verified")

        if has_money and amount >= 3000:
            print(f"✅ Has PKR {amount:,} — sufficient balance")
            print("\n🎉 LOAN APPROVED!")
        else:
            print("❌ Insufficient balance")
            print("\n❌ LOAN REJECTED — Balance too low")
    else:
        print("❌ No CNIC on file")
        print("\n❌ LOAN REJECTED — CNIC required")
else:
    print(f"❌ Age {age}: Must be at least 21")
    print("\n❌ LOAN REJECTED — Too young")
````

````python
# ── Multiple Conditions with and / or ─────────────────────────────────────

# Using 'and' — cleaner than deeply nested ifs
age        = 25
has_degree = True
gpa        = 3.5
experience = 2   # years

print("=== Job Application Screening ===")

if age >= 22 and has_degree and gpa >= 3.0:
    print("✅ Basic qualifications met")

    if experience >= 3:
        print("✅ Senior position eligible")
    elif experience >= 1:
        print("✅ Junior position eligible")
    else:
        print("⚠️  Entry-level only — no experience")
else:
    print("❌ Does not meet basic requirements")

print()

# Using 'in' with if — check if value is in a list
day = "Saturday"
weekend = ["Saturday", "Sunday"]

if day in weekend:
    print(f"{day} is a weekend — Office CLOSED 🏖️")
else:
    print(f"{day} is a weekday — Office OPEN 💼")
````

````python
# ── Ternary Operator (One-line if/else) ───────────────────────────────────
# value_if_true   if   condition   else   value_if_false
# Useful for simple assignments — not for complex logic!

age = 20

# Traditional way
if age >= 18:
    status = "Adult"
else:
    status = "Minor"
print(f"Traditional : {status}")

# Ternary (one-line) way
status = "Adult" if age >= 18 else "Minor"
print(f"Ternary     : {status}")

print()

# More examples
score = 75
result  = "Pass" if score >= 60 else "Fail"
label   = "Even" if score % 2 == 0 else "Odd"
discount = 0.15 if score >= 90 else 0.05

print(f"Score {score}: {result}, {label}, Discount: {discount*100:.0f}%")

# In f-strings directly!
for s in [45, 72, 88, 95, 55]:
    print(f"  Score {s:>3}: {'Pass ✅' if s >= 60 else 'Fail ❌'}")
````

# 🔁 SECTION 3: `for` Loops — Repeating Over Collections

## What Is a `for` Loop?

A `for` loop **repeats a block of code** for every item in a collection (list, string, range, dictionary, etc.).

**Real life analogy:**
```
FOR each student in the class:
    mark their attendance
    print their name
```

**Python syntax:**
```python
for variable in collection:
    # code to repeat
    # indented!
```

The **variable** takes each item's value, one at a time, on each iteration:
```python
for student in ["Ahmed", "Fatima", "Zara"]:
    print(student)
# 1st: student = "Ahmed"  → prints Ahmed
# 2nd: student = "Fatima" → prints Fatima
# 3rd: student = "Zara"   → prints Zara
```

### 3.1 — Looping Over Lists and Strings

````python
# ── for loop over a list ───────────────────────────────────────────────────

students = ["Ahmed", "Fatima", "Zara", "Hassan", "Sara"]

# Basic loop
print("=== All Students ===")
for student in students:
    print(f"  Hello, {student}!")

print()

# Loop with a calculation
prices = [1200, 850, 3500, 200, 4800]
print("=== Price with 10% Tax ===")
for price in prices:
    tax       = price * 0.10
    total     = price + tax
    print(f"  Price: PKR {price:>5,}  |  Tax: PKR {tax:>5,.0f}  |  Total: PKR {total:>6,.0f}")

print()

# Loop over a string (strings are sequences of characters!)
print("=== Characters in 'Python' ===")
for char in "Python":
    print(f"  '{char}'")
````

### 3.2 — The `range()` Function

````python
# ── range() — Generate a sequence of numbers ───────────────────────────────
# range(stop)             → 0 to stop-1
# range(start, stop)      → start to stop-1
# range(start, stop, step)→ start to stop-1, skipping by step

# range(5) → 0, 1, 2, 3, 4
print("range(5)        :", list(range(5)))

# range(2, 8) → 2, 3, 4, 5, 6, 7
print("range(2, 8)     :", list(range(2, 8)))

# range(1, 11, 2) → 1, 3, 5, 7, 9
print("range(1,11,2)   :", list(range(1, 11, 2)))

# range(10, 0, -1) → 10, 9, 8, ... 1 (countdown!)
print("range(10,0,-1)  :", list(range(10, 0, -1)))

print()

# Using range in a loop
print("=== Multiplication Table of 7 ===")
for i in range(1, 11):
    result = 7 * i
    print(f"  7 x {i:>2} = {result:>3}")

print()

# Use range to loop by index
students = ["Ahmed", "Fatima", "Zara"]
print("=== With Index ===")
for i in range(len(students)):
    print(f"  Index {i}: {students[i]}")
````

### 3.3 — `enumerate()` and `zip()`

````python
# ── enumerate() — Index + Value together ──────────────────────────────────
# Avoids needing a separate counter variable
# Returns (index, value) tuples

cities = ["Islamabad", "Lahore", "Karachi", "Peshawar", "Quetta"]

# Without enumerate (old way)
print("Without enumerate:")
for i in range(len(cities)):
    print(f"  {i}: {cities[i]}")

print()

# With enumerate (Pythonic way)
print("With enumerate:")
for index, city in enumerate(cities):
    print(f"  {index}: {city}")

print()

# Start counting from 1 instead of 0
print("Starting from 1:")
for number, city in enumerate(cities, start=1):
    print(f"  {number}. {city}")
````

````python
# ── zip() — Loop through TWO lists at once ────────────────────────────────
# Pairs up items from multiple lists
# Stops when the shortest list ends

names   = ["Ahmed",  "Fatima", "Zara",    "Hassan",  "Sara"]
scores  = [85,       92,       78,         95,        88]
cities  = ["Islamabad", "Lahore", "Karachi", "Peshawar", "Quetta"]

# Zip two lists
print("=== Student Scores ===")
for name, score in zip(names, scores):
    status = "Pass" if score >= 60 else "Fail"
    print(f"  {name:<10}: {score}  [{status}]")

print()

# Zip three lists
print("=== Full Student Info ===")
print(f"  {'NAME':<10} {'SCORE':>6} {'CITY':<12} {'GRADE':>6}")
print("  " + "-" * 38)
for name, score, city in zip(names, scores, cities):
    grade = "A" if score >= 90 else "B" if score >= 80 else "C"
    print(f"  {name:<10} {score:>6} {city:<12} {grade:>6}")
````

### 3.4 — `break`, `continue`, `pass`

````python
# ── break — Exit the loop immediately ─────────────────────────────────────
# When Python hits 'break', it stops the loop and jumps past it

print("=== break: Stop when Hassan is found ===")
students = ["Ahmed", "Fatima", "Zara", "Hassan", "Sara", "Bilal"]

for student in students:
    if student == "Hassan":
        print(f"  Found Hassan! Stopping search.")
        break
    print(f"  Checking: {student}")

print("Loop ended.")

print()

# Real use: find first failing student
print("=== Find first failing score ===")
scores = [88, 92, 75, 45, 67, 55]

for i, score in enumerate(scores):
    if score < 60:
        print(f"  First fail at index {i}: score = {score}")
        break
else:
    # The 'else' on a for loop runs ONLY if the loop completed without break
    print("  No failing scores found!")
````

````python
# ── continue — Skip this iteration, move to next ──────────────────────────
# 'continue' jumps back to the top of the loop immediately

print("=== continue: Skip absent students ===")
students = ["Ahmed", "Fatima", None, "Hassan", None, "Sara"]
#                              ↑ absent               ↑ absent

for student in students:
    if student is None:         # skip absent slots
        print("  [Absent]")
        continue                # skip rest of loop body
    print(f"  ✅ Present: {student}")

print()

# Real use: skip negative or invalid values
print("=== Sum only positive numbers ===")
data  = [10, -5, 8, -3, 15, -1, 7]
total = 0

for num in data:
    if num < 0:
        print(f"  Skipping {num}")
        continue
    total += num
    print(f"  Adding {num} → running total: {total}")

print(f"Final total: {total}")
````

````python
# ── pass — Placeholder (do nothing) ───────────────────────────────────────
# 'pass' is a placeholder — Python requires at least one line in a block
# Use it when you want to write the structure now but fill in code later

scores = [85, 92, 45, 78, 55, 91]

print("=== pass: Structure placeholder ===")
for score in scores:
    if score >= 90:
        print(f"  {score}: Top scorer!")
    elif score >= 60:
        pass   # TODO: add middle-range logic later
    else:
        print(f"  {score}: Needs improvement")

print()
print("(Middle range scores were silently skipped using pass)")
````

### 3.5 — Nested `for` Loops

````python
# ── Nested for loops ───────────────────────────────────────────────────────
# A loop inside another loop
# Outer loop runs N times
# For EACH outer iteration, inner loop runs M times
# Total iterations = N × M

# Example 1: Multiplication table grid
print("=== Multiplication Table (1–5) ===")
print("     ", end="")
for j in range(1, 6):
    print(f"{j:>5}", end="")
print()
print("     " + "─" * 25)

for i in range(1, 6):          # outer loop — rows
    print(f"{i:>3} |", end="")
    for j in range(1, 6):     # inner loop — columns
        print(f"{i*j:>5}", end="")
    print()                   # new line after each row

print()

# Example 2: Checking every student's subject scores
print("=== Subject Score Checker ===")
gradebook = [
    {"name": "Ahmed",  "scores": [85, 92, 45]},
    {"name": "Fatima", "scores": [90, 88, 75]},
    {"name": "Hassan", "scores": [55, 42, 38]},
]
subjects = ["Math", "English", "Python"]

for student in gradebook:                          # outer: each student
    print(f"\n  {student['name']}:")
    for subject, score in zip(subjects, student["scores"]):   # inner: each subject
        flag = "⚠️ FAILING" if score < 60 else "✅"
        print(f"    {subject:<10}: {score}  {flag}")
````

````python
# ── Looping through Dictionaries ──────────────────────────────────────────

inventory = {
    "Laptop"  : {"price": 85000, "stock": 15, "brand": "Dell"},
    "Mouse"   : {"price": 1500,  "stock": 80, "brand": "Logitech"},
    "Keyboard": {"price": 3500,  "stock": 45, "brand": "HP"},
    "Monitor" : {"price": 42000, "stock": 8,  "brand": "Samsung"},
    "USB Hub" : {"price": 2000,  "stock": 3,  "brand": "Anker"},
}

print("=" * 62)
print("                   📦 INVENTORY REPORT")
print("=" * 62)
print(f"  {'PRODUCT':<12} {'BRAND':<12} {'PRICE':>10} {'STOCK':>7} {'STATUS':<15}")
print("  " + "─" * 58)

total_value  = 0
low_stock    = []
out_of_stock = []

for product, details in inventory.items():
    price      = details["price"]
    stock      = details["stock"]
    brand      = details["brand"]
    item_value = price * stock
    total_value += item_value

    if stock == 0:
        status = "❌ OUT OF STOCK"
        out_of_stock.append(product)
    elif stock < 10:
        status = "⚠️  LOW STOCK"
        low_stock.append(product)
    else:
        status = "✅ OK"

    print(f"  {product:<12} {brand:<12} PKR {price:>7,} {stock:>7}  {status}")

print("  " + "─" * 58)
print(f"  Total Inventory Value: PKR {total_value:>10,}")
print("=" * 62)

if low_stock:
    print(f"\n⚠️  Reorder soon: {', '.join(low_stock)}")
if out_of_stock:
    print(f"❌  Out of stock: {', '.join(out_of_stock)}")
````

# 🔄 SECTION 4: `while` Loops — Repeating Until a Condition is Met

## What Is a `while` Loop?

A `while` loop keeps running **as long as a condition is `True`**.  
You use it when you **don't know in advance** how many times to repeat.

```python
while condition:
    # runs while condition is True
    # indented!
```

**Comparison:**

| | `for` loop | `while` loop |
|-|------------|---------------|
| Use when | You know the **count** or have a **collection** | You repeat until a **condition changes** |
| Example | Loop over all students in a list | Loop until user types "quit" |
| Risk | None | ⚠️ Infinite loop if condition never becomes False |

## ⚠️ The Infinite Loop Danger!

If the condition **never becomes False**, the loop runs forever and crashes your program!

```python
# ❌ INFINITE LOOP — never add this:
while True:
    print("This runs forever!")

# ✅ SAFE — always make sure the condition eventually becomes False:
count = 0
while count < 5:
    count += 1   # ← this changes count so loop eventually ends
```

### 4.1 — Basic `while` Loop

````python
# ── Basic while loop ───────────────────────────────────────────────────────

# Count from 1 to 5
print("=== Counting 1 to 5 ===")
count = 1                    # initialization — BEFORE the loop
while count <= 5:            # condition — checked BEFORE each iteration
    print(f"  Count: {count}")
    count += 1               # update — INSIDE the loop (critical!)
print("Done!")

print()

# Countdown
print("=== Countdown ===")
n = 5
while n > 0:
    print(f"  {n}...")
    n -= 1
print("  🚀 Launch!")

print()

# Sum numbers until sum exceeds a limit
print("=== Sum until > 100 ===")
total    = 0
number   = 1
while total <= 100:
    total  += number
    print(f"  Added {number:>3} → Total: {total}")
    number += 1
print(f"Stopped at total = {total} (exceeded 100)")
````

### 4.2 — `while` with `break` and `continue`

````python
# ── while with break ──────────────────────────────────────────────────────
# 'while True' + 'break' is a very common pattern
# The loop runs forever UNTIL it hits break

# Simulate PIN entry (without real input for classroom demo)
correct_pin = "1234"
attempts    = 0
max_attempts= 3
test_pins   = ["0000", "9999", "1234"]   # simulated user inputs

print("=== ATM PIN Verification ===")
for entered_pin in test_pins:
    attempts += 1
    print(f"  Attempt {attempts}: Entered '{entered_pin}'")

    if entered_pin == correct_pin:
        print("  ✅ PIN correct! Access granted.")
        break
    elif attempts >= max_attempts:
        print("  ❌ Too many failed attempts. Card blocked.")
        break
    else:
        print(f"  ❌ Wrong PIN. {max_attempts - attempts} attempt(s) left.")

print()

# while with continue — skip specific values
print("=== Print odd numbers from 1 to 10 ===")
num = 0
while num < 10:
    num += 1
    if num % 2 == 0:    # even number
        continue        # skip to next iteration
    print(f"  {num}")
````

````python
# ── Real use: Process a queue of tasks ────────────────────────────────────

task_queue = ["Send email", "Update database", "Generate report",
              "Call client", "Submit invoice"]

print(f"Tasks to process: {len(task_queue)}")
print("=" * 42)

task_number = 1
while task_queue:                          # while list is not empty
    current_task = task_queue.pop(0)       # take from front
    print(f"  [{task_number}] Processing: {current_task}")
    print(f"      Remaining tasks: {len(task_queue)}")
    task_number += 1

print("=" * 42)
print("✅ All tasks completed!")
````

````python
# ── while vs for — choosing the right tool ────────────────────────────────

# FOR loop — when you know what you're iterating over
marks = [85, 92, 78, 95, 88]

print("=== FOR: Process each mark ===")
total = 0
for mark in marks:       # you KNOW the collection
    total += mark
print(f"Total: {total}, Average: {total/len(marks):.2f}")

print()

# WHILE loop — when repetition depends on a condition, not a collection
print("=== WHILE: Keep adding until > 300 ===")
total    = 0
i        = 0
while total <= 300:      # you DON'T know how many iterations ahead of time
    total += marks[i % len(marks)]    # cycle through marks
    i     += 1
    print(f"  After iteration {i}: total = {total}")
print(f"Reached {total} after {i} iterations")
````

### 4.3 — The `while/else` and Avoiding Infinite Loops

````python
# ── while/else ────────────────────────────────────────────────────────────
# 'else' block runs ONLY if loop ended naturally (condition became False)
# Does NOT run if loop was terminated by 'break'

# Scenario: search for a student
roster   = ["Ahmed", "Fatima", "Zara", "Hassan", "Sara"]
search   = "Hassan"
index    = 0
found    = False

print(f"Searching for '{search}' in roster...")

while index < len(roster):
    if roster[index] == search:
        print(f"  ✅ Found '{search}' at position {index}!")
        found = True
        break
    print(f"  Checking index {index}: {roster[index]}")
    index += 1
else:
    print(f"  ❌ '{search}' was not found in the roster")

print()

# Guard against infinite loops — always have a max iteration safety
print("=== Safe loop with max iterations ===")
value    = 1
max_iter = 100    # safety limit
iteration = 0

while value < 1000:
    value    *= 2
    iteration += 1

    if iteration >= max_iter:
        print(f"  ⚠️  Safety limit reached at iteration {iteration}")
        break

print(f"  Value {value} reached in {iteration} iterations (doubled each time)")
````

````python
# ══════════════════════════════════════════════════════════════════
#  EXERCISE 2: Number Guessing Game
#  Topics: while loop + if/elif/else + break
# ══════════════════════════════════════════════════════════════════

"""
A guessing game where:
- A secret number is set (simulated, no real input for classroom)
- The player has 5 attempts
- After each guess: 'Too high', 'Too low', or 'Correct!'
- Track attempts used
- Show remaining attempts after each guess
"""

import random

secret_number = random.randint(1, 100)
max_attempts  = 5
attempts_used = 0

# Simulated guesses for classroom (replace with input() for real game)
# We'll generate guesses that demonstrate the game logic
guesses = []
low, high = 1, 100
for _ in range(max_attempts):
    guesses.append((low + high) // 2)     # binary search strategy
    mid = (low + high) // 2
    if mid == secret_number: break
    elif mid < secret_number: low = mid + 1
    else: high = mid - 1

print(f"(Secret number is: {secret_number} — binary search guesses: {guesses})")
print()
print("=" * 42)
print("       🎯 NUMBER GUESSING GAME")
print(f"       (Guess between 1 and 100)")
print("=" * 42)

game_won = False

while attempts_used < max_attempts:
    if attempts_used >= len(guesses):
        break

    guess         = guesses[attempts_used]
    attempts_used += 1
    remaining     = max_attempts - attempts_used

    print(f"\n  Attempt {attempts_used}: Guessing {guess}")

    if guess == secret_number:
        print(f"  🎉 CORRECT! You found {secret_number} in {attempts_used} attempt(s)!")
        game_won = True
        break
    elif guess < secret_number:
        print(f"  📉 Too LOW! Go higher.")
    else:
        print(f"  📈 Too HIGH! Go lower.")

    if remaining > 0:
        print(f"  {remaining} attempt(s) remaining.")

if not game_won:
    print(f"\n  ❌ Out of attempts! The number was {secret_number}.")

print("\n" + "=" * 42)
````

````python
# ══════════════════════════════════════════════════════════════════
#  EXERCISE 3: Sales Analysis Dashboard
#  Topics: Dictionary + for loop + if/elif/else + while
# ══════════════════════════════════════════════════════════════════

"""
Analyze monthly sales data:
1. Find the best and worst month
2. Calculate total and average monthly sales
3. Classify each month as High/Medium/Low sales
4. Find how many months to reach 500,000 total (cumulative)
5. Show a simple text bar chart
"""

monthly_sales = {
    "January"  : 82000,
    "February" : 73000,
    "March"    : 95000,
    "April"    : 110000,
    "May"      : 125000,
    "June"     : 140000,
    "July"     : 98000,
    "August"   : 88000,
    "September": 115000,
    "October"  : 132000,
    "November" : 158000,
    "December" : 175000,
}

# ─── Basic Statistics ───────────────────────────────────────────────
sales_values = list(monthly_sales.values())
total_sales  = sum(sales_values)
avg_sales    = total_sales / len(sales_values)
max_sales    = max(sales_values)
min_sales    = min(sales_values)

best_month  = [m for m, s in monthly_sales.items() if s == max_sales][0]
worst_month = [m for m, s in monthly_sales.items() if s == min_sales][0]

# Thresholds for classification
high_threshold = avg_sales * 1.15
low_threshold  = avg_sales * 0.85

print("=" * 58)
print("             📊 ANNUAL SALES DASHBOARD")
print("=" * 58)
print(f"  {'MONTH':<12} {'SALES':>10}  {'TREND':>8}  PERFORMANCE")
print("  " + "─" * 54)

prev_sales = None
for month, sales in monthly_sales.items():

    # Month-over-month trend
    if prev_sales is None:
        trend = "  —  "
    elif sales > prev_sales:
        change = ((sales - prev_sales) / prev_sales) * 100
        trend  = f"↑{change:>4.1f}%"
    elif sales < prev_sales:
        change = ((prev_sales - sales) / prev_sales) * 100
        trend  = f"↓{change:>4.1f}%"
    else:
        trend  = "  0%  "

    # Performance classification
    if sales >= high_threshold:
        perf = "🔥 HIGH"
    elif sales <= low_threshold:
        perf = "❄️  LOW"
    else:
        perf = "✅ MED"

    print(f"  {month:<12} PKR {sales:>8,}  {trend:>8}   {perf}")
    prev_sales = sales

print("  " + "─" * 54)
print(f"  {'Total':<12} PKR {total_sales:>8,}")
print(f"  {'Average':<12} PKR {avg_sales:>8,.0f}")
print("=" * 58)
print(f"  🏆 Best month  : {best_month} (PKR {max_sales:,})")
print(f"  📉 Worst month : {worst_month} (PKR {min_sales:,})")

# ─── Cumulative target using while ─────────────────────────────────
print()
target     = 500000
cumulative = 0
months_hit = 0

for month, sales in monthly_sales.items():
    cumulative += sales
    months_hit += 1
    if cumulative >= target:
        print(f"  🎯 Reached PKR {target:,} by end of {month} (month {months_hit})")
        break

# ─── Text bar chart ────────────────────────────────────────────────
print()
print("  📊 Sales Bar Chart (each █ = PKR 10,000)")
print("  " + "─" * 54)
for month, sales in monthly_sales.items():
    bars = "█" * (sales // 10000)
    print(f"  {month[:3]} | {bars} {sales//1000}K")
````

````python
# ══════════════════════════════════════════════════════════════════
#  EXERCISE 4: Word Frequency Counter
#  Topics: for loop + if + dictionary (building from scratch)
# ══════════════════════════════════════════════════════════════════

"""
Count how many times each word appears in a sentence.
This is a classic data analysis task!
"""

text = """
Python is a great language. Python is used for data analysis.
Data analysis is important in every field. Python makes data
analysis easy and fun. Every data analyst should learn Python.
"""

# Clean and split into words
import string
clean_text  = text.lower()
for char in string.punctuation:
    clean_text = clean_text.replace(char, " ")
words = clean_text.split()

print(f"Total words: {len(words)}")
print()

# Count frequencies using a dictionary
word_freq = {}

for word in words:
    if word in word_freq:
        word_freq[word] += 1     # word seen before — increment
    else:
        word_freq[word] = 1      # first time seeing this word

# Sort by frequency (highest first)
sorted_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

print("=" * 38)
print("  📊 WORD FREQUENCY REPORT")
print("=" * 38)
print(f"  {'WORD':<15} {'COUNT':>6}  BAR")
print("  " + "─" * 34)

for word, count in sorted_freq:
    bar = "█" * count
    print(f"  {word:<15} {count:>6}  {bar}")

print("=" * 38)
print(f"  Unique words: {len(word_freq)}")
print(f"  Most common : '{sorted_freq[0][0]}' ({sorted_freq[0][1]} times)")
````

````python
# ══════════════════════════════════════════════════════════════════
#  EXERCISE 5: FizzBuzz + Pattern (Classic Interview Question)
#  Topics: for loop + if/elif/else
# ══════════════════════════════════════════════════════════════════

"""
FizzBuzz rules:
- If number divisible by 3: print 'Fizz'
- If number divisible by 5: print 'Buzz'
- If divisible by BOTH 3 and 5: print 'FizzBuzz'
- Otherwise: print the number

Extension: also draw a number triangle
"""

print("=== FizzBuzz (1 to 30) ===")
result_line = []

for i in range(1, 31):
    if i % 15 == 0:       # divisible by BOTH 3 and 5 — check this FIRST!
        result_line.append("FizzBuzz")
    elif i % 3 == 0:
        result_line.append("Fizz")
    elif i % 5 == 0:
        result_line.append("Buzz")
    else:
        result_line.append(str(i))

# Print in rows of 5
for i in range(0, len(result_line), 5):
    row = result_line[i:i+5]
    print("  " + "  ".join(f"{item:>8}" for item in row))

print()

# Number triangle pattern
print("=== Number Triangle ===")
rows = 6
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()

print()

# Star pyramid
print("=== Star Pyramid ===")
size = 6
for i in range(1, size + 1):
    spaces = " " * (size - i)
    stars  = "* " * i
    print(spaces + stars)
````

## 📝 Day 23 Summary

### Dictionaries:

| Concept | Syntax | Remember |
|---------|--------|---------|
| Create | `{"key": value}` | Curly braces, colon separates key & value |
| Access | `d["key"]` or `d.get("key")` | Use `.get()` to avoid KeyError |
| Add/Update | `d["key"] = value` | Same syntax for both |
| Remove | `del d["key"]` or `d.pop("key")` | `pop()` returns the value |
| Loop | `for k, v in d.items()` | `.items()` gives key-value pairs |
| Comprehension | `{k: v for k, v in ...}` | Like list comprehension |

### Conditionals (`if/elif/else`):

| Concept | Remember |
|---------|----------|
| Syntax | Ends with `:`, body must be **indented** |
| `elif` | Checked only if previous `if` was False |
| `else` | Runs when ALL conditions above are False |
| `and` | BOTH conditions must be True |
| `or` | AT LEAST ONE must be True |
| `not` | Reverses the boolean |
| Ternary | `value_a if condition else value_b` |

### `for` Loops:

| Concept | Syntax |
|---------|--------|
| Basic | `for item in collection:` |
| With counter | `for i in range(n):` |
| With index | `for i, v in enumerate(list):` |
| Two lists | `for a, b in zip(list1, list2):` |
| Stop early | `break` |
| Skip item | `continue` |

### `while` Loops:

| Concept | Remember |
|---------|----------|
| Use when | You don't know iteration count — depends on a condition |
| Always | Initialize before, update inside — or you get infinite loop! |
| `break` | Exit immediately |
| `continue` | Skip to next iteration |
| `while/else` | `else` runs only if loop ended naturally (not via break) |
