# Matplotlib Basics: Figure, Axes, and the Pyplot API

Matplotlib offers two primary interfaces:

- Pyplot state-machine (`plt.plot`, etc.) for quick scripts
- Object-oriented API (`fig, ax = plt.subplots()`) for control and composition


```python
x = np.arange(-100, 100)
x[-10:]
```

```python
squared = np.square(x)
sin_values = np.sin(x)
cos_values = np.cos(x)
cubed = np.power(x,3)
```

```python
# line plot
plt.plot(x, squared)
plt.title("Square Function")
plt.xlabel("x values (0-99)")
plt.ylabel('Squared values')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 7))
# Axes 0, 0
ax[0,0].plot(x, squared, c="red")
ax[0, 0].set(title="Square Function")

# Axes 0, 1
ax[0,1].plot(x, cubed, c = 'blue')
ax[0,1].set(title="Cube Function")

# Axes 1, 0
ax[1, 0].plot(x, sin_values)
ax[1, 0].set(title="Sin Function")

# Axes 0, 1
ax[1,1].plot(x, cos_values)
ax[1, 1].set(title="Cos Function")


fig.suptitle("Mathematical functions")
plt.show()
```

## Histogram

```python
fig, ax = plt.subplots()
ax.hist(height, bins=20)
ax.set(title="Height distribution", xlabel="Height (inches)", ylabel="Count")
plt.show()
```

```python
## Box plot - more informative
fig, ax = plt.subplots()
ax.boxplot(height, vert=True, patch_artist=True,
           boxprops=dict(facecolor='lightblue', color='navy'),
           medianprops=dict(color='red', linewidth=2),
           whiskerprops=dict(color='yellow'),
           capprops=dict(color='red'),
           flierprops=dict(marker='o', color='green', alpha=1.0))
ax.set(
    title="Height Distribution (Box Plot)",
    ylabel="Height (inches)",
    xticks=[1],
    xticklabels=["Sample Heights"]
)
ax.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Qualitative/Categorical Data Plots
Categorical data are discrete labels or groups.
## Bar plot


```python
# Create synthetic categorical data for bar plot demonstration
categories = ["A", "B", "C", "D"]
counts = [23, 45, 12, 37]

fig, ax = plt.subplots()
ax.bar(categories, counts, color="skyblue")
ax.set(title="Product Category Counts", xlabel="Category", ylabel="Count")

# # Annotate bars with their values
for bar in ax.patches:
    height = bar.get_height()
    ax.annotate(f'{height}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 1),  # 1 point vertical offset
                textcoords="offset points",
                ha = 'center'
               )

plt.show()
```

```python
male = small_hw[small_hw['Gender'] == 'Male']['Height'].reset_index(drop=True)
female = small_hw[small_hw['Gender'] == 'Female']['Height'].reset_index(drop=True)

gender_hw = pd.DataFrame(
    {
        'Male Height': male,
        "Female Height": female
    },
    index=range(len(female))
)
gender_hw
```

```python
# Fill NaN with Mean / Median
mean = gender_hw['Male Height'].mean()
median = gender_hw['Male Height'].median()

print('Mean: ', mean)
print('Median: ', median)
```

```python
gender_hw['Male Height'] = gender_hw['Male Height'].fillna(mean)
```

```python
gender_hw['Male Height'].isnull().sum()
```

```python
gender_hw
```

```python
gender_hw.columns.values
```

```python
male_stats = gender_hw['Male Height'].max()
female_stats = gender_hw['Female Height'].max()
```

```python

genders = gender_hw.columns.values
stats = [male_stats, female_stats]

fig, ax = plt.subplots()
ax.bar(genders, stats, color="skyblue")
ax.set(title="Gender based Height Stats", xlabel="Gender", ylabel="Stats")

# # Annotate bars with their values
for bar in ax.patches:
    height = bar.get_height()
    ax.annotate(f'{height}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 1),  # 1 point vertical offset
                textcoords="offset points",
                ha = 'center'
               )

plt.show()
```

## Pie chart

```python
# Create synthetic categorical data for bar plot demonstration
categories = ["A", "B", "C", "D"]
counts = [23, 45, 12, 37]

# Pie chart for product category counts
fig, ax = plt.subplots()
ax.pie(counts, labels=categories, autopct='%10.0f%%', startangle=180, colors=plt.cm.Pastel1.colors)
ax.set(title="Product Category Distribution")
ax.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle
plt.show()
```
